```python
def updateObservation(json_data, observation_data, i, patientID, observation):
    observation_update = observation
    observation_update["resource"]["valueQuantity"]["value"] = observation_data['VALUE'].values[i]
    observation_update["resource"]["valueQuantity"]["unit"] = observation_data['UNITS'].values[i]
    observation_update["resource"]["valueQuantity"]["code"] = observation_data['UNITS'].values[i]
    observation_update["resource"]["code"]["coding"][0]["code"] = observation_data['CODE'].values[i]
    observation_update["resource"]["code"]["coding"][0]["display"] = observation_data['DESCRIPTION'].values[i]
    observation_update["resource"]["code"]["text"] = observation_data['DESCRIPTION'].values[i]
    observation_update["resource"]["effectiveDateTime"] = observation_data['DATE'].values[i]
    observation_update["resource"]["issued"] = observation_data['DATE'].values[i]
    observation_update["resource"]["subject"]["reference"] = patientID
    json_data["entry"].append(observation_update)
```
